```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data series
w = np.arange(20, 20000, 10)

# Define constants
C = 45.847e-9
R = 1.4748e+3

# Define the transfer function
def H(w):
    return (1j * w * C * R) / (1 + ((1j * w * C * R)))

# Define the phase angle function
def P(w):
    return np.arctan(1 / (w * C * R)) * (180 / np.pi)

# Define the frequencies measured at
measured_w = np.array([
    200,
    1000,
    2000,
    3000,
    4000,
    5000,
    6000,
    7000,
    8000,
    9000,
    10000,
    11000,
    12000,
    13000,
    14000,
    15000,
    16000,
    17000,
    18000,
    19000,
    20000,
])

# Define the measured transfer function ratios
measured_H = np.array([
    0.0975,
    0.405,
    0.6525,
    0.785,
    0.865,
    0.899014778,
    0.928921569,
    0.948780488,
    0.951923077,
    0.961904762,
    0.969194313,
    0.974056604,
    0.974238876,
    0.981220657,
    0.983490566,
    0.978672986,
    0.983050847,
    0.98019802,
    0.984575835,
    0.991913747,
    0.988636364,
])

# Define the measured phase angles
measured_P = np.array([
    80,
    64,
    48,
    38,
    30,
    25,
    22,
    19,
    16,
    15,
    13,
    12,
    11,
    10,
    9,
    8,
    7,
    7,
    7,
    6,
    5,
])
```

```python
# Plot the magnitude of the transfer function against frquency
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (16, 8)
plt.plot(w, np.absolute(H(w)))
plt.plot(measured_w, measured_H)
plt.ylabel("Magnitude")
plt.xlabel("Frequency (Hz)")
plt.show()
```

```python
# Plot the phase angle against frquency
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (16, 8)
plt.plot(w, P(w))
plt.plot(measured_w, measured_P)
plt.ylabel("Phase angle (Degrees)")
plt.xlabel("Frequency (Hz)")
plt.show()
```
