```python
import numpy as np
import matplotlib
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

# Define the data series
f = np.arange(20, 20000, 10)

# Define constants
C = 45.847e-9
R = 1.4748e+3

def w(f):
    return 2 * np.pi * f

# Define the transfer function
def H(f):
    return (1j * w(f) * C * R) / (1 + ((1j * w(f) * C * R)))

# Define the phase angle function
def P(f):
    return np.arctan(1 / (w(f) * C * R)) * (180 / np.pi)

# Define the cut-off frequency
cof = 1 / (2 * np.pi * C * R)

# Define the frequencies measured at
measured_f = np.array([
    200,
    1000,
    2000,
    3000,
    4000,
    5000,
    6000,
    7000,
    8000,
    9000,
    10000,
    11000,
    12000,
    13000,
    14000,
    15000,
    16000,
    17000,
    18000,
    19000,
    20000,
])

# Define the measured transfer function ratios
measured_H = np.array([
    0.0975,
    0.405,
    0.6525,
    0.785,
    0.865,
    0.899014778,
    0.928921569,
    0.948780488,
    0.951923077,
    0.961904762,
    0.969194313,
    0.974056604,
    0.974238876,
    0.981220657,
    0.983490566,
    0.978672986,
    0.983050847,
    0.98019802,
    0.984575835,
    0.991913747,
    0.988636364,
])

# Define the measured phase angles
measured_P = np.array([
    80,
    64,
    48,
    38,
    30,
    25,
    22,
    19,
    16,
    15,
    13,
    12,
    11,
    10,
    9,
    8,
    7,
    7,
    7,
    6,
    5,
])
```

```python
# Plot the magnitude of the transfer function against frequency
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (16, 8)
theoretical = plt.plot(f, np.absolute(H(f)))
measured = plt.plot(measured_f, measured_H)
plt.axvline(x=cof, alpha=0.3)
plt.axhline(y=(1/np.sqrt(2)), alpha=0.3)
plt.ylabel("Magnitude")
plt.xlabel("Frequency (Hz)")
keys = [
    mpatches.Patch(color='blue', label='Theoretical'),
    mpatches.Patch(color='orange', label='Measured'),
]
plt.legend(handles=keys)
plt.show()
```
